# **2. Exploratory data analysis**

## 2.1 Univariate Analysis

### 2.1.a  Density plot of change in population per each county

```python
fig, ax = plt.subplots(figsize=(20,10))        
sns.distplot(data.PST120214)
plt.title("Density plot of population change per county")
plt.xlabel("Population change from 2010 to 2014")
plt.ylabel("Density")
plt.show()
```

#### Insights:
population change is between -10 to 20 and right skewed. The population of some counties has decreased from 2010 to 2014 and most of the counties shows an increase in population.

## 2.2 Bivariate Analysis

```python
dem_votes=pd.DataFrame(data.groupby("PostalAbbr")["Democrat"].sum())
```

## 2.2.2  Line plots showing votes aggregation among Blacks and Whites population

```python
#plot of republican votes among Blacks population
plt.figure(figsize=(15,8))
sns.lineplot(data['RHI225214'],data['Republican'],color='red')
plt.title('Republican votes VS Blacks percent', size=15)
plt.show()
```

#### Insights:
   larger number of republican votes were from counties where percentage of Black/African Americans in the population is around 23%.

```python
#plot of democrat votes among blacks
plt.figure(figsize=(15,8))
sns.lineplot(data['RHI225214'],data['Democrat'],color='blue')
plt.title('Democrat votes VS Blacks percent', size=20)
plt.show()
```

#### findings:
   larger number of democrat votes were in counties where percentage of Black/African Americans is around 8%.

```python
#plot of republican votes among whites population
plt.figure(figsize=(15,8))
sns.lineplot(data['RHI125214'],data['Republican'],color='red')
plt.title('Republican votes VS whites percent', size=20)
plt.show()
```

####  findings:
most of the republican votes are concentrated on counties where white population are about 75%.
    

```python
#plot of democrat votes among whites
plt.figure(figsize=(15,8))
sns.lineplot(data['RHI125214'],data['Democrat'],color='blue')
plt.title('Democrat votes VS Whites percent', size=20)
plt.show()
```
